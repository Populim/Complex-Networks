The average clustering coefficient and the transitivity of the following 4 networks will be analysed:<br>
The Euroroad Network:http://konect.cc/networks/subelj_euroroad<br>
The Facebook Network:http://konect.cc/networks/subelj_euroroad<br>
The Human Proteins Network:http://konect.uni-koblenz.de/networks/maayan-vidal<br>
The Petster Hamster's Friendship Network:http://konect.uni-koblenz.de/networks/petster-friendships-hamster<br>
All networks are unweighted and undirected.<br>

Importing essential libraries:

```python
import matplotlib.pyplot as plt
import networkx as nx
import collections
from numpy import *
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
G = nx.read_edgelist("euroroad.txt")

pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size = 10)
plt.show()

print("Average Clustering: ", nx.average_clustering(G))
print("Transitivity: ", nx.transitivity(G))
```
