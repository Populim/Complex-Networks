The cumulative degree distribution of the following 4 networks will be analysed:<br>
The Euroroad Network:http://konect.cc/networks/subelj_euroroad<br>
The Facebook Network:http://konect.cc/networks/subelj_euroroad<br>
The Human Proteins Network:http://konect.uni-koblenz.de/networks/maayan-vidal<br>
The Petster Hamster's Friendship Network:http://konect.uni-koblenz.de/networks/petster-friendships-hamster<br>
All networks are unweighted and undirected.<br>

Importing essential libraries:

```python
import matplotlib.pyplot as plt
import networkx as nx
import collections
from numpy import *
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
def plotCumulativeDegreeDistribution(G):
    #extracting the biggest connected component:
    Gc = max(nx.connected_component_subgraphs(G), key=len)
    
    #adding the degree count
    degree_sequence = sorted([Gc.degree(n) for n in Gc.nodes()])
    degreeCount = collections.Counter(degree_sequence)
    deg, cnt = zip(*degreeCount.items())
    cumulative = np.cumsum(cnt)
    cumulative = cumulative[::-1]

    #plotting the distribution
    plt.step(deg,cumulative)
    plt.yscale('log')
    plt.xscale('log')
    plt.show()
    return
```

```python
G = nx.read_edgelist("euroroad.txt")
pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size = 10)
```

```python
plotCumulativeDegreeDistribution(G)
```

```python
G = nx.read_edgelist("facebookEx5.txt")
pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size = 5)
```

```python
plotCumulativeDegreeDistribution(G)
```

```python
G = nx.read_edgelist("vidal_protein.txt")
pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size = 5)
```

```python
plotCumulativeDegreeDistribution(G)
```

```python
G = nx.read_edgelist("petsterhamsters.txt")
pos = nx.spring_layout(G)
nx.draw(G, pos = pos, node_size = 5)
```

```python
plotCumulativeDegreeDistribution(G)
```

None of them are scale free, as a scale free network would follow a power law distribution, which would show in the plots as a straight line.
