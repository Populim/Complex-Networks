All networks are unweighted and undirected.<br>

Importing essencial libraries:

```python
import matplotlib.pyplot as plt
import networkx as nx
import collections
from numpy import *
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import time
```

```python
G = nx.read_edgelist("euroroad.txt")
pos = nx.spring_layout(G)

nx.draw(G, pos = pos, node_size = 10)
plt.show()
```

Calculating the average shortest path length:

```python

#extracting the biggest connected component from the network
Gc = max(nx.connected_component_subgraphs(G), key=len)

value = nx.average_shortest_path_length(Gc)
print("Average shortest path length of the network: ", value)
```

```python

G = nx.read_edgelist("facebookEx5.txt")

#extracting the biggest connected component from the network
Gc = max(nx.connected_component_subgraphs(G), key=len)

value = nx.average_shortest_path_length(Gc)
print("Average shortest path length of the network: ", value)
```

```python
#http://konect.uni-koblenz.de/networks/petster-friendships-hamster
G = nx.read_edgelist("petsterhamsters.txt")

#extracting the biggest connected component from the network 
Gc = max(nx.connected_component_subgraphs(G), key=len)

value = nx.average_shortest_path_length(Gc)
print("Average shortest path length of the network: ", value)
```
