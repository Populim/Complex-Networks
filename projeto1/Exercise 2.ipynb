Importing essencial libraries:

```python
import networkx as nx
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
```

Reading the Les Miserables network and plotting its adjacency matrix(AM):

```python
G =  nx.read_edgelist('lesmis.txt', nodetype=int,
  data=(('weight',int),), create_using=nx.DiGraph())

matrix =  nx.to_numpy_matrix(G)
plt.imshow(matrix, cmap='PRGn', interpolation='none')
plt.colorbar()
plt.show()
```

Plotting the network:

```python
pos = nx.spring_layout(G)
nx.draw(G, pos=pos,node_size = 30, width = 0.2)
```

Transforming the network undirected:

```python
G = G.to_undirected()
#pos = nx.spring_layout(G)
nx.draw(G, pos=pos, node_size = 15, width = 0.1)
```

Turning the network unweighted:

```python
matrix =  nx.to_numpy_matrix(G)

matrix[matrix > 0] = 1 #if an element of the matrix is not zero, give it the value 1
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

Now, let's do the same with the Foodweb Network:

```python
G = nx.read_edgelist("eco-foodweb-baywet.edges", nodetype=int,
                    data=(('weight',float),), create_using = nx.DiGraph())
pos = nx.spring_layout(G)
nx.draw(G,pos = pos, node_size = 15, width = 0.1)
```

Plotting the AM to vizualise the weights:

```python
matrix = nx.to_numpy_matrix(G)
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

We can see the weights differ in a big range.<br>
Let's unweight and undirect this network as well:<br>

```python
matrix[matrix !=0] = 1
G = nx.from_numpy_matrix(matrix)
G = nx.to_undirected(G)
matrix = nx.to_numpy_matrix(G)
plt.imshow(matrix)
plt.colorbar()
plt.show()
```

Plotting the transformed network:

```python
pos = nx.spring_layout(G)
nx.draw(G,pos = pos, node_size = 15, width = 0.1)
```
